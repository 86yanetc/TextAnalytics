## Preparing Data to pass to model

```python
import numpy as np
index = df.index
df['random_number'] = np.random.randn(len(index))
train = df[df['random_number'] <= 0.8]
test = df[df['random_number'] > 0.8]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(token_pattern=r'\b\w+\b')
train_matrix = vectorizer.fit_transform(train['translated'])
test_matrix = vectorizer.transform(test['translated'])
```

## Training Model on Data

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
X_train = train_matrix
X_test = test_matrix
y_train = train['sentiment']
y_test = test['sentiment']
```

```python
lr.fit(X_train,y_train)
```

```python
predictions = lr.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
new = np.asarray(y_test)
confusion_matrix(predictions,y_test)
```

## Printing results of training

```python
print(classification_report(predictions,y_test))
```

```python
count=0
for prediction, true_value in zip(predictions,y_test):
    if prediction == true_value:
        count+=1
print('accuracy is', round((count/len(predictions))*100,2), '%.')
```

```python
import pandas as pd
import spacy
```

```python
spacy.load('en_core_web_sm')
from spacy.lang.en import English
parser = English()
def tokenize(text):
    lda_tokens = []
    tokens = parser(text)
    for token in tokens:
        if token.orth_.isspace():
            continue
        elif token.like_url:
            lda_tokens.append('URL')
#         elif token.orth_.startswith('@'):
#             lda_tokens.append('SCREEN_NAME')
        else:
            lda_tokens.append(token.lower_)
    return lda_tokens
```

```python
import nltk
nltk.download('wordnet')
from nltk.corpus import wordnet as wn
def get_lemma(word):
    lemma = wn.morphy(word)
    if lemma is None:
        return word
    else:
        return lemma
    
from nltk.stem.wordnet import WordNetLemmatizer
def get_lemma2(word):
    return WordNetLemmatizer().lemmatize(word)
```

```python
nltk.download('stopwords')
en_stop = set(nltk.corpus.stopwords.words('english'))
```

```python
def prepare_text_for_lda(text):
    tokens = tokenize(text)
    tokens = [token for token in tokens if len(token) > 4]
    tokens = [token for token in tokens if token not in en_stop]
    tokens = [token for token in tokens if token != 'https']
    tokens = [get_lemma(token) for token in tokens]
    tokens = [get_lemma(token) if token not in ('corona', 'covid', 'covid19', 'covid-19', 'coronavirus', 'virus') else 'covid' for token in tokens]
    return tokens
```

```python
import random
text_data = []
def tokens_(line):
    tokens = prepare_text_for_lda(line)
    text_data.append(tokens)
    if random.random() > .99:
        print(tokens)
```

```python
_ = df['translated'].apply(tokens_)
```

```python
from gensim import corpora
dictionary = corpora.Dictionary(text_data)
corpus = [dictionary.doc2bow(text) for text in text_data]
import pickle
pickle.dump(corpus, open('corpus.pkl', 'wb'))
dictionary.save('dictionary.gensim')
```

```python
import gensim
NUM_TOPICS = 5
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics = NUM_TOPICS, id2word=dictionary, passes=15)
ldamodel.save('model5.gensim')
topics = ldamodel.print_topics(num_words=10)
for topic in topics:
    print(topic)
```
